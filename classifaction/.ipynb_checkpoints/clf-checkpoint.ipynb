```python
#引入模块
import os
import scipy.io as scio
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
import pandas
minMax = MinMaxScaler()
```

```python
#处理数据
data_dict = {}
data_path = "./data"
for i in os.listdir(data_path):
    data = scio.loadmat(os.path.join(data_path,i))
    data_dict[i.split(".")[0]] = data[i.split(".")[0]]
```

```python
for i in data_dict.keys():
    data_dict[i] = minMax.fit_transform(data_dict[i])
    
    
```

```python
X = []
for i in data_dict.keys():
    for j in data_dict[i]:
        X.append(j)
X = np.array(X)
```

```python
Y = []
count = 0
label = 0
for i in data_dict.keys():
    for j in range(len(data_dict[i])):
        Y.append(label)
        count += 1
    label += 1
```

```python
Y = np.array(Y)
```

```python
Y
```

```python
idx = np.random.randint(0,6924,6924)
```

```python
#随机打乱
X,Y = X[idx],Y[idx]
```

```python
#训练
clf = SVC()
clf.fit(X,Y)
```

```python
#预测
y_pred = clf.predict(X)
```

```python
valid = (Y == y_pred)
```

```python
test_path = './test.mat'
test_data = scio.loadmat(test_path)['data_test_final']
```
